## Desenvolupament amb HST

```python
import hostphot.local_photometry as lp

from astropy.cosmology import WMAP9 as cosmo

# Calcular kpc por arcsec
kpc_per_arcsec = cosmo.kpc_proper_per_arcmin(z).value / 60.0  # convertir a kpc/arcsec

# Ahora, para definir ap_radii para una apertura de 1 kpc:
ap_radii = [1 / kpc_per_arcsec]  # 1 kpc convertido a arcsec usando kpc_per_arcsec

results = lp.multi_band_phot(name, ra, dec, z, filters=filt,
                             survey=survey, ap_radii=ap_radii, correct_extinction=True,
                             save_plots=True, raise_exception=True,use_mask=False)
results
```
