```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import hostphot
from hostphot.cutouts import download_images
from hostphot.coadd import coadd_images
from hostphot.image_masking import create_mask
from hostphot.utils import plot_fits, get_survey_filters
import hostphot.local_photometry as lp
import hostphot.global_photometry as gp
from hostphot.cutouts import set_HST_image
```

# Cutouts

```python
from hostphot.cutouts import set_HST_image

name = 'IEPO2V010'
ra, dec =  19.67456,-7.44498
host_ra, host_dec = 19.678548,-7.450020   # coords of host galaxy
z = 0.017752  # redshift

survey = 'HST'
filt = 'WFC3_UVIS_F275W'

file = 'hst_16741_2v_wfc3_uvis_f275w_iepo2v_drc.fits'
set_HST_image(file, filt, name)
```

## Desenvolupament amb HST

```python
import hostphot.local_photometry as lp

from astropy.cosmology import WMAP9 as cosmo

# Calcular kpc por arcsec
kpc_per_arcsec = cosmo.kpc_proper_per_arcmin(z).value / 60.0  # convertir a kpc/arcsec

# Ahora, para definir ap_radii para una apertura de 1 kpc:
ap_radii = [1 / kpc_per_arcsec]  # 1 kpc convertido a arcsec usando kpc_per_arcsec

results = lp.multi_band_phot(name, ra, dec, z, filters=filt,
                             survey=survey, ap_radii=ap_radii, correct_extinction=True,
                             save_plots=True, raise_exception=True)
results
```

## Desenvolupament amb PS1

```python
from hostphot.cutouts import download_images

name = 'ASASSN-13AR'
ra, dec =  19.67456,-7.44498
host_ra, host_dec = 19.678548,-7.450020   # coords of host galaxy
z = 0.017752  # redshift
#surveys = ['HST']
surveys = ['PS1']
for survey in surveys:
    download_images(name, host_ra, host_dec,
                    survey=survey, overwrite=True)
```

```python
fits_files = []
for survey in surveys:
    survey_files = glob.glob(f'images/{name}/{survey}*')
    # Utilitzant la funció glob.glob, el codi busca fitxers en el directori images/{name}/ 
    #que comencin amb el nom de l'enquesta (survey) i tinguin qualsevol extensió (l'asterisc indica això).
    for file in survey_files:
        if os.path.basename(file).startswith(survey) and file.endswith('.fits'):
            #Es comprova que el nom del fitxer comença amb el nom de l'enquesta i acaba amb l'extensió .fits. 
            #Si compleix aquestes condicions, el fitxer s'afegeix a la llista fits_files
            fits_files.append(file)
fits_files = list(np.sort(fits_files))
#Finalment, els fitxers FITS es posen en ordre alfabètic utilitzant np.sort() (de la llibreria numpy), 
#i es converteixen en una llista de nou.
```

```python
for file in fits_files:
    # Per a cada fitxer, es comprova si el nom del fitxer conté alguna de les següents subcadenes
    if '_r.fits' in file or '_NUV.fits' in file or '_J.fits' in file or '_W1.fits' in file:
        plot_fits(file)
```

# Coadds

```python
#El diccionari coadd_dict associa cadascuna d’aquestes enquestes astronòmiques amb els seus filtres:
coadd_dict = {
              'PS1':'riz'
             }
for survey, coadd_filters in coadd_dict.items():
    coadd_images(name, filters=coadd_filters, survey=survey)

#Per a cada enquesta (survey) i els seus filtres associats (coadd_filters), es crida la funció coadd_images(), 
#que és responsable de generar imatges coadicionades (sumar o combinar dades d'imatges d'aquests filtres en la mateixa enquesta)
```

# Masking

```python
for survey, coadd_filters in coadd_dict.items():
    filters = get_survey_filters(survey)
    #Se llama a la función get_survey_filters(survey) para obtener todos los filtros disponibles de la encuesta actual (survey). 
    #Esto devuelve una lista de filtros específicos para la encuesta (filters).
    coadd_mask_params = create_mask(name, host_ra, host_dec,
                                    filt=coadd_filters,
                                    survey=survey,
                                    extract_params=True)

    for filt in filters: #Creación de máscaras para cada filtro individual
        create_mask(name, host_ra, host_dec,
                    filt, survey=survey,
                    common_params=coadd_mask_params)
        #common_params=coadd_mask_params: Esto asegura que la máscara de cada filtro individual se construya utilizando los mismos 
        #parámetros de la máscara de la imagen coadicionada, manteniendo la coherencia entre todas las máscaras
```

# Plot masks

```python
masked_files = []
for survey in surveys:
    survey_files = glob.glob(f'images/{name}/masked_{survey}*')
    # S’utilitza la funció glob.glob() per cercar tots els fitxers dins de la carpeta images/{name}/ 
    #que compleixin el patró masked_{survey}*
    for file in survey_files:
        if survey in file and file.endswith('.fits'):
        # El nom de l'enquesta (survey) ha d'estar dins del nom del fitxer i El fitxer ha de tenir l'extensió .fits
            masked_files.append(file)

masked_files = list(np.sort(masked_files)) #ordenar
```

```python
for file in masked_files:
    if '_r.fits' in file or '_NUV.fits' in file or '_J.fits' in file or '_W1.fits' in file:
        plot_fits(file)

        jpg_file = file.replace('.fits', '.jpg')
        plt.figure(figsize=(20, 12))
        img = mpimg.imread(jpg_file)
        imgplot = plt.imshow(img)
        plt.show()
```

# Local photometry

```python
from astropy.cosmology import WMAP9 as cosmo

# Redshift
z = 0.0178

# Calcular kpc por arcsec
kpc_per_arcsec = cosmo.kpc_proper_per_arcmin(z).value / 60.0  # convertir a kpc/arcsec

# Ahora, para definir ap_radii para una apertura de 1 kpc:
ap_radii = [1 / kpc_per_arcsec]  # 1 kpc convertido a arcsec usando kpc_per_arcsec

for survey in surveys:
    # los resultados se guardarán en diferentes archivos
    result = lp.multi_band_phot(name, ra, dec, z,
                       survey=survey, ap_radii=ap_radii,
                       use_mask=True, correct_extinction=True,
                       save_plots=True, save_results=True,
                       raise_exception=True)
    print(f"Resultados para {survey}: {result}")






'''S'obtenen resultats per a cada banda:
g_5, r_5, i_5, z_5, y_5 : Magnituds mesurades a les bandes g (verd), r (vermell), i (infraroig proper), z (infraroig llunyà) i 
i (infraroig proper de banda ampla), respectivament.

g_5_flux, r_5_flux, i_5_flux, z_5_flux, y_5_flux: Aquests són els fluxos (flux) mesurats a cada banda

g_zeropoint, r_zeropoint, i_zeropoint, z_zeropoint, y_zeropoint: Aquests són els punts zero (zeropoint) utilitzats a cada banda.'''
```
